```python
%matplotlib inline
import benfordslaw as bl
import benford as bf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import locale
from locale import atof
plt.style.use('ggplot')
```

```python
locale.setlocale(locale.LC_NUMERIC, '')
```

```python
siconv_proposta=pd.read_csv('siconv_proposta.csv',encoding='UTF-8',sep=';')
```

```python
siconv_emenda=pd.read_csv('siconv_emenda.csv',encoding='UTF-8',sep=';')
```

```python
siconv = siconv_emenda.set_index('ID_PROPOSTA').join(siconv_proposta.set_index('ID_PROPOSTA'))
```

```python
eduardo_siconv = siconv.loc[siconv['NOME_PARLAMENTAR'] == 'EDUARDO BOLSONARO']
```

```python
jean_siconv = siconv.loc[siconv['NOME_PARLAMENTAR'] == 'JEAN WYLLYS']
```

```python
jean_valores = pd.DataFrame(jean_siconv['VL_GLOBAL_PROP'])
```

```python
jean_valores_tratado = jean_valores.apply(lambda x: pd.to_numeric(x.astype(str).str.replace(',',''), errors='coerce'))
```

```python
benford_jean = bl.fit(jean_valores_tratado['VL_GLOBAL_PROP'])
```

```python
bl.plot(benford_jean, title='Jean Wyllys')
```

```python
bf.first_digits(jean_valores_tratado['VL_GLOBAL_PROP'], digs=1, decimals=2)
```
